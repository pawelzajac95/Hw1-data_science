```python
import numpy as np

print('')
def positive_numbers():

    a = np.arange(1, 11)
    print('Zadanie 1.')
    print('===========')
    return a

def matrix_3x3():

    a = np.zeros((3, 3), dtype=int)
    print('\nZadanie 2.')
    print('============')
    return a

def rand_5x5():

    a = np.random.randint(low=1, high=11, size=(5, 5))    
    print('\nZadanie 3.')
    print('============')
    return a

def rand_4x4():

    a = np.random.rand(4, 4)
    print('\nZadanie 4.')
    print('============')
    return a

def two_array():

    a = np.random.randint(low=1, high=11, size=5)
    b = np.random.randint(low=1, high=11, size=5)
    sum = a + b
    sub = a - b
    mul = a * b
    print('\nZadanie 5.')
    print('============')
    print('Array 1:\n', a)
    print('Array 2:\n', b)
    print('Sum:\n', sum)
    print('Sub:\n', sub)
    print('Mul:\n', mul)

    return ''
    
def two_vectors():

    a = np.random.rand(7)
    b = np.random.rand(7)
    scalar_sub = np.dot(a, b)
    print('\nZadanie 6.')
    print('============')
    print('Vector 1:\n', a)
    print('Vector 2:\n', b)    
    return scalar_sub

def two_matrix():

    a = np.random.randint(low=1, high=11, size=(2, 2))
    b = np.random.randint(low=1, high=11, size=(2, 3))
    result = np.dot(a, b)
    print('\nZadanie 7.')
    print('============')
    print('Matrix 1:\n', a)
    print('Matrix 2:\n', b)
    return result

def reciprocal():

    a = np.random.randint(low=1, high=11, size=(3, 3))
    result = np.linalg.inv(a)
    print('\nZadanie 8.')
    print('============')
    print('Matrix:\n', a)
    return result

def trans():
    
    a = np.random.random((4, 4))
    result = a.T
    print('\nZadanie 9.')
    print('============')
    print('Matrix:\n', a)
    return result

def matrix_3x4():

    a = np.random.randint(low=1, high=11, size=(3, 4))
    b = np.random.randint(low=1, high=11, size=4)
    result = a * b
    print('\nZadanie 10.')
    print('=============')
    print("Matrix:\n", a)
    print('Vector:\n', b)
    return result

def matrix_2x3():

    a = np.random.random((2, 3))
    b = np.random.random(3)
    result = a * b
    print('\nZadanie 11.')
    print('=============')
    print("Matrix:\n", a)
    print('Vector:\n', b)
    return result

def matrix_2x2():

    a = np.random.randint(low=1, high=11, size=(2, 2))
    b = np.random.randint(low=1, high=11, size=(2, 2))
    result = a * b
    print('\nZadanie 12.')
    print('=============')
    print("Matrix 1:\n", a)
    print('Matrix 2:\n', b)
    return result

def matrix_2x2_2():

    a = np.random.randint(low=1, high=11, size=(2, 2))
    b = np.random.randint(low=1, high=11, size=(2, 2))
    result = np.dot(a, b)
    print('\nZadanie 13.')
    print('=============')
    print("Matrix 1:\n", a)
    print('Matrix 2:\n', b)
    return result

def matrix_5x5():

    a = np.random.randint(low=1, high=101, size=(5, 5))
    result = np.sum(a)
    print('\nZadanie 14.')
    print('=============')
    print("Matrix:\n", a)
    return result

def matrix_4x4():

    a = np.random.randint(low=1, high=11, size=(4, 4))
    b = np.random.randint(low=1, high=11, size=(4, 4))
    result = a - b
    print('\nZadanie 15.')
    print('=============')
    print("Matrix 1:\n", a)
    print('Matrix 2:\n', b)
    return result

def matrix_3x3_16():

    a = np.random.random((4, 4))
    result = np.sum(a, axis=1).reshape(-1, 1)
    print('\nZadanie 16.')
    print('=============')
    print('Matrix:\n', a)
    
    return result

def matrix_square():
    
    a = np.array([[2, 5, 8, 1], [1, 6, 3, 4], [7, 0, 9, 2]])
    result = np.square(a)
    print('\nZadanie 17.')
    print('=============')
    print('Matrix:\n', a)   
    return result

def vector_square():

    a = np.random.randint(low=1, high=51, size=4)
    result = np.sqrt(a)
    print('\nZadanie 18.')
    print('=============')
    print('Matrix:\n', a)   
    return result

print('Result:\n', positive_numbers())
print('Result:\n', matrix_3x3())
print('Result:\n', rand_5x5())
print('Result:\n', rand_4x4())
print(two_array())
print('Result:\n', two_vectors())
print('Result:\n', two_matrix())
print('Result:\n', reciprocal())
print('Result:\n', trans())
print('Result:\n', matrix_3x4())
print('Result:\n', matrix_2x3())
print('Result:\n', matrix_2x2())
print('Result:\n', matrix_2x2_2())
print('Result:\n', matrix_5x5())
print('Result:\n', matrix_4x4())
print('Result:\n', matrix_3x3_16())
print('Result:\n', matrix_square())
print('Result:\n', vector_square())
```
